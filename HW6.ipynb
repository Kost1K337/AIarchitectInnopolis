```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/drive/MyDrive/Архитектор ИИ/Titanic-Dataset.csv')
```

```python
df = df.dropna()
```

```python
#df.groupby('Age').Survived.hist()
df['age group'] = pd.cut(df.Age, bins=range(0, 100, 10), right=False)
df
```

```python
groups = df[df['Survived'] == 1].groupby('age group')
for i, (name, group) in enumerate(groups):
  plt.hist(group['Survived']+i, label = 'Survived'  if i == 0 else "", color = 'b')
groups = df[df['Survived'] == 0].groupby('age group')
for i, (name, group) in enumerate(groups):
  plt.hist(group['Survived']+i+.1, label = 'Deceased' if i == 0 else "", color = 'r')
plt.legend()
plt.xticks(np.arange(0.1,9.6,1),[f"({i*10}, {i*10+10}]" for i in range(10)], fontsize=7)
plt.title('Histogram of survaval by age groups')
plt.xlabel('Age')
```
